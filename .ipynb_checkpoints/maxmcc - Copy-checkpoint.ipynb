```python
import numpy as np
import pandas as pd

#bdtnp = pd.read_csv('data/binarized_bdtnp.csv')
bdtnp = pd.read_csv('data/bdtnp.csv')
#dge = pd.read_csv('data/dge_binarized_distMap.csv').T.reset_index(drop=True)
dge = pd.read_csv('data/magic_dge.csv')
a2g = pd.read_csv('data/area_to_geometry.csv')
g2a = pd.read_csv('data/gene_to_area.csv')

dict_g2a = {}

for col in g2a.columns:
    for val in g2a[col]:
        dict_g2a[val] = col
#'btd': 'Ant',
#'Idgf1': 'Ven',
#'betaTub60D': 'Ant',
#'wor': 'Ven',

dict_a2g = {'Ant':[], 'Dor':[], 'Lat':[], 'Pos':[], 'Ven':[]}
for index, row in a2g.iterrows():
    dict_a2g[row['V4']].append(index)
```

```python
gene_areas_counter = {'Ant':0, 'Dor':0, 'Lat':0, 'Pos':0, 'Ven':0}
for col in bdtnp.columns:
    if(col in dict_g2a):
        gene_areas_counter[dict_g2a[col]] += 1

gene_areas_counter
```

Sum of all genes is 401 hence the weights for scoring regions are:

Ant: 401/96, Dor: 401/44, Lat:401/41, Pos:401/144, Ven:401/76

```python
areas = []

for index, row in bdtnp.iterrows():
    gene_areas_counter = {'Ant':0, 'Dor':0, 'Lat':0, 'Pos':0, 'Ven':0}
    for i in range(0, len(row)):
        if(row.index[i] in dict_g2a):
            gene_areas_counter[dict_g2a[row.index[i]]] += row[i]
    
    gene_areas_counter['Ant'] *= (44/9) #(401/96)
    gene_areas_counter['Dor'] *= (44/2) #(401/44)
    gene_areas_counter['Lat'] *= (44/7) #(401/41)
    gene_areas_counter['Pos'] *= (44/19) #(401/144)
    gene_areas_counter['Ven'] *= (44/7) #(401/76)
    #print(gene_areas_counter)
    areas.append(max(gene_areas_counter, key=gene_areas_counter.get))
"""
128up            1.524224
140up            1.275327
14-3-3epsilon    8.565557
"""
```
